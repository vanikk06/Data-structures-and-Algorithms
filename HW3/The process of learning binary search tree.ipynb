```python
def insert(root, val):
    if root == None:
        return

    if val != None:
        if val <= root.val:
            if root.left == None:
                root.left = TreeNode(val)                
            else:
                insert(root.left, val)                    
        else:
            if root.right == None:
                root.right = TreeNode(val)                
            else:
                insert(root.right, val)
    
    return TreeNode(val)
```
