## idea

ä¸€é–‹å§‹ä¸€ç›´è¢«ä¹‹å‰åƒè€ƒç¶²ç«™çš„Codeç¶ä½ï¼ŒåŸæœ¬æƒ³èªªé‚£ä¹¾è„†ä¸è¦ç”¨arrayï¼Œè©¦è©¦çœ‹ç”¨linked liståšåšçœ‹quick sortï¼Œçµæœç™¼ç¾æœ‰æŠ€è¡“ä¸Šçš„å›°é›£ğŸ˜‚ï¼Œå¾Œä¾†éˆå…‰ä¸€é–ƒï¼Œæƒ³åˆ°å¯ä»¥æŠŠlinked liståŸ·è¡Œquick sortçš„æ–¹æ³•ï¼Œç”¨arrayåŸ·è¡Œ

     Â§ linked listæ¶æ§‹quick sortæƒ³æ³•ï¼šå°‡æœ€å¾Œä¸€å€‹å€¼è¨­ç‚ºkeyï¼Œå¾é ­é–‹å§‹å¾€å¾Œæ‰¾ï¼Œæ‰¾åˆ°å¤§æ–¼keyçš„å€¼å°±å¾€å¾Œä¸Ÿ

```python
def quick_sort(list):
    if len(list) <= 1:                    #å…ˆåˆ¤æ–·listæ˜¯å¦æ˜¯å¯ä»¥æ¯”å¤§å°çš„
        return list
    
    bigger = []                          #å‰µé€ ä¸€å€‹ç©ºé–“æ”¾å¤§æ–¼keyçš„å€¼ï¼Œæœ€å¾Œåœ¨ä¸€èµ·å¾€å¾ŒåŠ 
    
    key = list[-1]                       #å°‡æœ€å¾Œä¸€å€‹å€¼è¨­ç‚ºkey
    i = 0
    while i < len(list):                 #ä¸€å€‹å€‹èˆ‡keyæ¯”è¼ƒ
        if list[i] <= key:               #<=keyå°±æ”¾è‘—ä¸å‹•ï¼Œå¾€ä¸‹ä¸€å€‹ç¹¼çºŒæ¯”
            i += 1
        else:                            #>keyå°±æŠ“å‡ºä¾†æ”¾åˆ°å‰›å‰›å»ºç«‹çš„ç©ºé–“ï¼Œä¸¦æŠŠå®ƒåœ¨åŸæœ¬çš„listä¸Šåˆªé™¤
            bigger.append(list.pop(i))      #å› ç‚ºåŸæœ¬çš„å€¼è¢«åˆªé™¤ï¼Œindexæ”¹è®Šï¼Œæ‰€ä»¥iä¸åŠ 1
    
    key_list = []                        #æŠŠæª¢æŸ¥éå¾Œçš„keyæŠ“å‡ºä¾†å¦å¤–æ”¾ï¼Œä¸¦å¾åŸæœ¬çš„listä¸Šåˆªé™¤
    key_list.append(list[-1])            
    list.pop()
    
    list = quick_sort(list)              #éè¿´å¼åŸ·è¡Œ
    bigger = quick_sort(bigger)
    return list+key_list+bigger
```

```python
list = [3,2,15,0,3,4,2,4,7]
quick_sort(list)
```
